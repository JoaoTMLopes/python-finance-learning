Faz um gráfico com matplotlib comparando a evolução normalizada dos preços (dica: divide cada série pelo primeiro valor, para todos começarem em 1.0 e seres possível comparar performance relativa).

```python
import matplotlib.pyplot as plt
import yfinance as yf
```

```python
ativos= ["GALP.LS", "BCP.LS", "JMT.LS"]
dados_psi_3 = yf.download(ativos, start="2023-01-01", end="2026-06-30")["Close"]
print(dados_psi_3.head())
```

```python
retornos_diarios = dados_psi_3.pct_change()
print(retornos_diarios)
resumo_estatistico = retornos_diarios.describe()
print(resumo_estatistico)
```

```python
precos_normalizados = dados_psi_3 / dados_psi_3.iloc[0]

plt.figure(figsize=(12, 6))
plt.plot(precos_normalizados.index, precos_normalizados["GALP.LS"], label="Galp")
plt.plot(precos_normalizados.index, precos_normalizados["BCP.LS"], label="BCP")
plt.plot(precos_normalizados.index, precos_normalizados["JMT.LS"], label="Jerónimo Martins")
plt.title("Evolução Normalizada dos Preços ", fontsize=14, fontweight="bold")
plt.ylabel("Performance Relativa (Base 1.0)", fontsize=12)
plt.xlabel("Data", fontsize=12)
```
