```python
import matplotlib.pyplot as plt
import yfinance as yf
```

```python
ativos= ["GALP.LS", "BCP.LS", "JMT.LS"]
dados_psi_3 = yf.download(ativos, start="2023-01-01", end="2026-06-30")["Close"]
print(dados_psi_3.head())
```

```python
retornos_diarios = dados_psi_3.pct_change()
print(retornos_diarios)
resumo_estatistico = retornos_diarios.describe()
print(resumo_estatistico)
```

```python
precos_normalizados = dados_psi_3 / dados_psi_3.iloc[0]

plt.figure(figsize=(12, 6))
plt.plot(precos_normalizados.index, precos_normalizados["GALP.LS"], label="Galp")
plt.plot(precos_normalizados.index, precos_normalizados["BCP.LS"], label="BCP")
plt.plot(precos_normalizados.index, precos_normalizados["JMT.LS"], label="Jerónimo Martins")
plt.title("Evolução Normalizada dos Preços ", fontsize=14, fontweight="bold")
plt.ylabel("Performance Relativa (Base 1.0)", fontsize=12)
plt.xlabel("Data", fontsize=12)
```

Passo3_Limpeza_Normalização_Visualização

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

ativos = ["MSFT", "AAPL", "SPY"]
dados = yf.download(ativos, start="2021-01-01", end="2026-06-30")["Close"]
print(dados.head())
```

```python
dados.isna().sum()
dados = dados.dropna()
```

```python
dados_normalizados = (dados / dados.iloc[0]) * 100
print(dados_normalizados.head())
```

```python
dados_normalizados.plot(figsize=(10,6), title="Performance Normalizada (Base 100)")
plt.xlabel("Data")
plt.ylabel("Índice (Base 100)")
plt.legend(title="Ativo")
plt.show()
```

```python
retornos = dados.pct_change().dropna()
resumo = pd.DataFrame({
    "Retorno_Medio_Diario": retornos.mean(),
    "Volatilidade_Diaria": retornos.std(),
    "Retorno_Total_Periodo": (dados.iloc[-1] / dados.iloc[0]) - 1
})
print(resumo)
```
