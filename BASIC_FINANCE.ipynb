```python
# 1. Criando as variáveis (Tipos de dados: Números e Inteiros)
preco_compra = 50.00      # Preço que você pagou pela ação (Float / Número decimal)
preco_atual = 58.50       # Preço que a ação vale hoje (Float / Número decimal)
quantidade = 100          # Quantidade de ações que você comprou (Int / Número inteiro)

# 2. Criando variáveis de texto (Strings) e booleanos para praticar
nome_da_acao = "PETR4"    # Texto sempre fica entre aspas (String)
em_lucro = preco_atual > preco_compra  # Guarda True ou False (Booleano)

# 3. Fórmula para calcular o lucro ou prejuízo
# Calculamos a diferença de preço por ação e multiplicamos pela quantidade total
resultado_total = (preco_atual - preco_compra) * quantidade

# 4. Mostrando os resultados na tela
print("--- RELATÓRIO DA CARTEIRA ---")
print("Ação:", nome_da_acao)
print("Quantidade:", quantidade)
print("Resultado Total: R$", resultado_total)
print("A operação está em lucro?", em_lucro)
```

```python
compra = { "Apple" : 100 , "S&P500": 125 , "MSFT" : 80 }
atual = {"Apple": 150 , "S&P500": 115 , "MSFT": 90}

# Adicionamos o 'reverse=True' no final do sorted()
compra_decrescente = dict(sorted(compra.items(), key=lambda item: item[1], reverse=True))
atual_decrescente = dict(sorted(atual.items(), key=lambda item: item[1], reverse=True))

print("Compra (Maior para Menor):", compra_decrescente)
print("Atual (Maior para Menor):", atual_decrescente)
```

```python
precos_antigos = [100, 200, 300]
precos_com_inflacao = []

for preco in precos_antigos:
    novo_preco = preco * 1.05  # Aumentando 5%
    precos_com_inflacao.append(novo_preco)

print(precos_com_inflacao)
# Resultado: [105.0, 210.0, 315.0]
```
