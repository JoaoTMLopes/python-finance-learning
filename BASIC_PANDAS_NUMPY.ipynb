```python
precos = np.array([100, 102, 98, 105, 110])
print(precos)
```

OPERAÇÕES_VETORIZADAS
Em Python, a indexação começa em 0, não em 1 (isto é uma das maiores diferenças e fontes de erro para quem vem de R). O primeiro elemento de precos é precos[0], não precos[1].
precos[1:] significa "todos os elementos a partir da posição 1 até ao fim" (ou seja, exclui o primeiro elemento).
precos[:-1] significa "todos os elementos exceto o último".
A subtração e divisão são feitas elemento a elemento automaticamente (vetorização), tal como em R quando fazes vetor1 - vetor2.
O resultado é o array de retornos percentuais diários.

```python
retorno = (precos[1:] - precos[:-1]) / precos[:-1]
print(retorno)
```

FUNÇÕES_ESTATISTICAS_BASICAS
Quando tamos no PYTHON eles usam o desvio padrao populacional (geral), se quisermos que seja igual ao R onde é trabalhado de acord com uma amostra tremos de addicionar  ddof=1

```python
np.mean(precos)      # equivalente a mean(precos) em R
np.std(precos)       # equivalente a sd(precos) em R
np.max(precos)       # equivalente a max(precos) em R
np.min(precos)       # equivalente a min(precos) em R
np.sum(precos)       # equivalente a sum(precos) em R
print (np.mean(precos))
print (np.std(precos))

np.std(precos, ddof=1)
print (np.std(precos, ddof=1))
```

EXERCICIO_PRATICO
Sem consultar exemplos, tenta sozinho:

Cria um DataFrame com 3 colunas: "Data" (8 datas fictícias em formato texto), "Ativo" (alterna entre dois nomes de ações/ETFs), "Preco" (valores fictícios, insere propositadamente 1-2 valores em falta usando None).

Converte a coluna "Data" para datetime e cria uma coluna nova "Mes" extraída da data.

Usa groupby("Ativo") para calcular o preço médio de cada ativo.

Detecta quantos valores em falta existem com .isna().sum() e preenche-os com a média da coluna.

Cria um segundo DataFrame pequeno com uma coluna "Setor" associada a cada ativo, e junta-o ao primeiro com pd.merge().

```python
import pandas as pd
import numpy as np

# 1. Criar o DataFrame original com 3 colunas e valores em falta (None)
dados = {
    "Data": [
        "2026-01-15", "2026-01-18", "2026-02-05", "2026-02-20",
        "2026-03-02", "2026-03-12", "2026-04-01", "2026-04-15"
    ],
    "Ativo": ["AAPL", "VWCE", "AAPL", "VWCE", "AAPL", "VWCE", "AAPL", "VWCE"],
    "Preco": [185.5, 112.0, None, 114.5, 190.2, None, 195.0, 113.8]
}

df = pd.DataFrame(dados)
print("--- DataFrame Inicial ---")
print(df, "\n")

# 2. Converter "Data" para datetime e extrair o "Mes"
df["Data"] = pd.to_datetime(df["Data"])
df["Mes"] = df["Data"].dt.month

# 3. Usar groupby("Ativo") para calcular o preço médio de cada ativo
medias_por_ativo = df.groupby("Ativo")["Preco"].mean()
print("--- Preço Médio por Ativo ---")
print(medias_por_ativo, "\n")

# 4. Detetar valores em falta e preenchê-los com a média global da coluna
valores_em_falta = df["Preco"].isna().sum()
print(f"Quantidade de valores em falta antes do preenchimento: {valores_em_falta}")

media_global = df["Preco"].mean()
df["Preco"] = df["Preco"].fillna(media_global)

print("\n--- DataFrame após preencher valores em falta ---")
print(df, "\n")

# 5. Criar um segundo DataFrame para o Setor e fazer o pd.merge()
dados_setor = {
    "Ativo": ["AAPL", "VWCE"],
    "Setor": ["Tecnologia", "Global ETF"]
}
df_setores = pd.DataFrame(dados_setor)

df_final = pd.merge(df, df_setores, on="Ativo", how="left")

print("--- DataFrame Final (com Merge) ---")
print(df_final)
```

```python
import pandas as pd

df = pd.read_csv("precos_ativos_semana10.csv")
df.head()
```

```python
df["Data"] = pd.to_datetime(df["Data"])
df["Preco"].isna().sum()
df = df.dropna(subset=["Preco"])
print(df.dropna(subset=["Preco"]))
```

```python
df["Data"] = pd.to_datetime(df["Data"])
df["Preco"] = df.groupby("Ativo")["Preco"].transform(lambda x: x.interpolate(method="linear"))
print(df.groupby)
```

```python
df = df.sort_values(["Ativo", "Data"])
df["Retorno_Diario"] = df.groupby("Ativo")["Preco"].pct_change()

df["Mes"] = df["Data"].dt.to_period("M")
retorno_mensal = df.groupby(["Ativo", "Mes"])["Retorno_Diario"].apply(lambda x: (1 + x).prod() - 1)
retorno_mensal
```

```python
import matplotlib.pyplot as plt

for ativo in df["Ativo"].unique():
    subset = df[df["Ativo"] == ativo]
    plt.plot(subset["Data"], subset["Preco"], label=ativo)

plt.legend()
plt.title("Evolução de Preços - VWCE vs AAPL")
plt.xlabel("Data")
plt.ylabel("Preço")
plt.show()
```
